```python
for col in df.columns:
  print(col,df[col].nunique())
  print(col,df[col].unique())
```

```python
x = df.drop(columns=['decision'])
y = df['decision']
```

```python
hoe = OneHotEncoder(drop='first',sparse=False)
x_features = hoe.fit_transform(x)
x_encoded = pd.DataFrame(x_features,columns=hoe.get_feature_names_out(x.columns))
x_encoded
```

```python
x_train,x_test,y_train,y_test= train_test_split(x_encoded,y,test_size=0.27,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred.shape
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
